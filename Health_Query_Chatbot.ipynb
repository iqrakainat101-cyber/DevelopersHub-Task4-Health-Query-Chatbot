```python
# Import Libraries
import google.generativeai as genai
import re
```

```python
# Add API Key
API_KEY = "YOUR_GEMINI_API_KEY"
genai.configure(api_key=API_KEY)
```

```python
# Check Available Models
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

```python
# Create the Model
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
# Safety Filter
def safety_check(user_query):

    dangerous_keywords = [
        "prescribe",
        "dosage",
        "suicide",
        "kill myself",
        "self harm",
        "overdose",
        "medical diagnosis",
        "which medicine should i take",
        "give me treatment"
    ]

    query_lower = user_query.lower()

    for word in dangerous_keywords:
        if word in query_lower:
            return False

    return True
```

```python
# Prompt Engineering Function
def health_chatbot(user_query):

    if not safety_check(user_query):
        return """
⚠️ Safety Notice:

I cannot provide medical diagnosis,
prescriptions, treatment plans,
or advice that could be harmful.

Please consult a qualified healthcare professional.
"""

    prompt = f"""
You are a helpful medical assistant.

Rules:
1. Provide general educational health information only.
2. Keep explanations simple and friendly.
3. Do NOT diagnose diseases.
4. Do NOT prescribe medications.
5. Do NOT provide dosage recommendations.
6. Encourage users to consult healthcare professionals for serious symptoms.
7. Use clear and easy language.

User Question:
{user_query}
"""

    response = model.generate_content(prompt)

    return response.text
```

```python
# Interactive Chatbot
print("="*60)
print("GENERAL HEALTH QUERY CHATBOT")
print("="*60)

while True:

    user_input = input("\nAsk a health-related question (or type 'exit'): ")

    if user_input.lower() == "exit":
        print("Chatbot Closed.")
        break

    answer = health_chatbot(user_input)

    print("\nBot Response:")
    print(answer)
```
